```python
#establish capture
cap = cv2.VideoCapture(os.path.join('Video','Aditya_Sample.mp4'))
#Loop[ through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    
    ret, frame = cap.read()
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    #show image
    cv2.imshow('Video Player', gray)
    
    #Breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

#close everything
cap.release()
cv2.destroyAllWindows()
```

```python
#Writing Out Videos
#Establish Capture
cap = cv2.VideoCapture(os.path.join('Video','Aditya_Sample.mp4'))

#setup Video Writer 

#Properties
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = cap.get(cv2.CAP_PROP_FPS)

#VideoWriter
video_writer = cv2.VideoWriter(os.path.join('Video','Output.avi'), cv2.VideoWriter_fourcc('X','V','I','D'), fps, (width, height), isColor=True)

#establish capture
cap = cv2.VideoCapture(os.path.join('Video','Aditya_Sample.mp4'))

#Loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    
    #Read Frame
    ret, frame = cap.read()

    #Gray transform
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    #show image
    cv2.imshow('Video Player', gray)

    #write out frames
    video_writer.write(frame)
    
    #Breaking out of the loop
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

#close everything
cap.release()
cv2.destroyAllWindows()

#Close or release video writer
video_writer.release()
```

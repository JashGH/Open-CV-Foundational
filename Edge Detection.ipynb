Import Dependencies

```python
#import opencv for computer vision
import cv2
#import os for file paths
import os
#from matplotlib import pyplot as plt
from matplotlib import pyplot as plt
```

Load Images

```python
IMG_PATH = os.path.join('dog.jpg')
IMG_PATH
img = cv2.imread(IMG_PATH)
# img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
#testing the size of the image
# (int(img.shape[1]/4), int(img.shape[0]/2.5))
```

```python
#Resizee Image
resized_img = cv2.resize(img, (int(img.shape[1]/4), int(img.shape[0]/4)))
#view Image using OpenCV
cv2.imshow('Frame View', resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

    Edge Detection

```python
#Apply canny
gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
blur = cv2.GaussianBlur(gray,(5,5), 0)
canny = cv2.Canny(blur, threshold1 = 200, threshold2 = 250)
```

```python
#Resizee Image
resized_img = cv2.resize(canny, (int(img.shape[1]/4), int(img.shape[0]/4)))
#view Image using OpenCV
cv2.imshow('Frame View', resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Export Edges

```python
cv2.imwrite('edges.jpg', canny)
```
